```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import time

# Lists to store data
titles = []
prices = []
furnishing_list = []
bathroom_list = []
tenant_list = []
availability_list = []
area_list = []
area_type_list = []
floor_no_list = []
total_floor_list = []
facing_list = []
balcony_list = []
descriptions = []
page_numbers = []

headers = {
    "content": "image/jpg",
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/145.0.0.0 Safari/537.36",
    "accept-encoding":"gzip, deflate, br, zstd"
    }

total_time = time.time()

for i in range(0, 51):

    start_time = time.time()

    URL = f"https://www.magicbricks.com/flats-for-rent-in-hyderabad-pppfr/page-{i}"

    page = requests.get(URL, headers=headers)
    soup = BeautifulSoup(page.text, "html.parser")

    cards = soup.find_all("div", class_="mb-srp__card__info mb-srp__card__info-withoutburger")

    for card in cards:

        # ------------------ TITLE ------------------
        title_tag = card.find("h2", class_="mb-srp__card--title")
        titles.append(title_tag.text.strip() if title_tag else np.nan)

        # ------------------ PRICE ------------------
        price_tag = card.find("div", class_="mb-srp__card__price")
        prices.append(price_tag.text.strip() if price_tag else np.nan)

        # ------------------ DESCRIPTION ------------------
        desc_tag = card.find("span", class_="remove-ellipsis")
        descriptions.append(desc_tag.text.strip() if desc_tag else np.nan)

        # ------------------ FEATURES ------------------
        furnishing = np.nan
        bathroom = np.nan
        tenant = np.nan
        availability = np.nan
        area = np.nan
        area_type = np.nan
        floor_no = np.nan
        total_floor = np.nan
        facing = np.nan
        balcony = np.nan

        summary_items = card.find_all("div", class_="mb-srp__card__summary__list--item")

        for item in summary_items:

            key = item.get("data-summary")
            value_tag = item.find("div", class_="mb-srp__card__summary--value")

            if not value_tag:
                continue

            value = value_tag.text.strip()

            if key == "furnishing":
                furnishing = value

            elif key == "bathroom":
                bathroom = value

            elif key == "tenent-preffered":
                tenant = value

            elif key == "status":
                availability = value

            elif key in ["super-area", "carpet-area"]:
                area = value.replace(" sqft", "")
                area_type = key

            elif key == "floor":
                if "out of" in value:
                    parts = value.split(" out of ")
                    floor_no = parts[0]
                    total_floor = parts[1]

            elif key == "facing":
                facing = value

            elif key == "balcony":
                balcony = value

        furnishing_list.append(furnishing)
        bathroom_list.append(bathroom)
        tenant_list.append(tenant)
        availability_list.append(availability)
        area_list.append(area)
        area_type_list.append(area_type)
        floor_no_list.append(floor_no)
        total_floor_list.append(total_floor)
        facing_list.append(facing)
        balcony_list.append(balcony)
        page_numbers.append(i)

    print(f"Page {i} completed in {time.time() - start_time:.2f} seconds")

    time.sleep(5)  # polite delay to avoid blocking

print("Total Time Completed in seconds:", time.time() - total_time)

# ------------------ DATAFRAME ------------------

df = pd.DataFrame({
    "Title": titles,
    "Price": prices,
    "Furnishing": furnishing_list,
    "Bathrooms": bathroom_list,
    "Tenant Preferred": tenant_list,
    "Availability": availability_list,
    "Area (sqft)": area_list,
    "Area Type": area_type_list,
    "Floor Number": floor_no_list,
    "Total Floors": total_floor_list,
    "Facing": facing_list,
    "Balcony": balcony_list,
    "Page Number": page_numbers
})

print(df.head())
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import time
import re

# Lists
titles = []
bhk_list = []
location_list = []
prices = []
furnishing_list = []
bathroom_list = []
tenant_list = []
availability_list = []
area_list = []
area_type_list = []
floor_no_list = []
total_floor_list = []
facing_list = []
balcony_list = []
page_numbers = []

headers = {
    "User-Agent": "Mozilla/5.0"
}

total_time = time.time()

for i in range(1, 51):

    start_time = time.time()

    URL = f"https://www.magicbricks.com/flats-for-rent-in-hyderabad-pppfr/page-{i}"
    page = requests.get(URL, headers=headers)
    soup = BeautifulSoup(page.text, "html.parser")

    cards = soup.find_all("div", class_="mb-srp__card__info mb-srp__card__info-withoutburger")

    for card in cards:

        # ---------------- TITLE ----------------
        title_tag = card.find("h2", class_="mb-srp__card--title")
        title = title_tag.text.strip() if title_tag else np.nan
        titles.append(title)

        # -------- Extract BHK from Title --------
        bhk_match = re.search(r'(\d+)\s*BHK', title) if title else None
        bhk = bhk_match.group(1) if bhk_match else np.nan
        bhk_list.append(bhk)

        # -------- Extract Location from Title --------
        # Format: "2 BHK Flat for Rent in Bachupally, Hyderabad"
        location = np.nan
        if title and " in " in title:
            location_part = title.split(" in ")[1]
            location = location_part.replace(", Hyderabad", "").strip()
        location_list.append(location)

        # ------------- PRICE from Description -------------
        desc_tag = card.find("div", class_="mb-srp__card--desc--text")
        description = desc_tag.text.strip() if desc_tag else ""

        price_match = re.search(r'₹\s?[\d,]+', description)
        price = price_match.group().replace("₹", "").replace(",", "") if price_match else np.nan
        prices.append(price)

        # ------------- FEATURES -------------
        furnishing = np.nan
        bathroom = np.nan
        tenant = np.nan
        availability = np.nan
        area = np.nan
        area_type = np.nan
        floor_no = np.nan
        total_floor = np.nan
        facing = np.nan
        balcony = np.nan

        summary_items = card.find_all("div", class_="mb-srp__card__summary__list--item")

        for item in summary_items:

            key = item.get("data-summary")
            value_tag = item.find("div", class_="mb-srp__card__summary--value")

            if not value_tag:
                continue

            value = value_tag.text.strip()

            if key == "furnishing":
                furnishing = value

            elif key == "bathroom":
                bathroom = value

            elif key == "tenent-preffered":
                tenant = value

            elif key == "status":
                availability = value

            elif key in ["super-area", "carpet-area"]:
                area = value.replace(" sqft", "")
                area_type = key

            elif key == "floor":
                if "out of" in value:
                    parts = value.split(" out of ")
                    floor_no = parts[0]
                    total_floor = parts[1]

            elif key == "facing":
                facing = value

            elif key == "balcony":
                balcony = value

        furnishing_list.append(furnishing)
        bathroom_list.append(bathroom)
        tenant_list.append(tenant)
        availability_list.append(availability)
        area_list.append(area)
        area_type_list.append(area_type)
        floor_no_list.append(floor_no)
        total_floor_list.append(total_floor)
        facing_list.append(facing)
        balcony_list.append(balcony)
        page_numbers.append(i)

    print(f"Page {i} completed in {time.time() - start_time:.2f} seconds")
    time.sleep(2)

print("Total Time:", time.time() - total_time)

# ---------------- CREATE DATAFRAME ----------------

df = pd.DataFrame({
    "Title": titles,
    "BHK": bhk_list,
    "Location": location_list,
    "Price": prices,
    "Furnishing": furnishing_list,
    "Bathrooms": bathroom_list,
    "Tenant Preferred": tenant_list,
    "Availability": availability_list,
    "Area (sqft)": area_list,
    "Area Type": area_type_list,
    "Floor Number": floor_no_list,
    "Total Floors": total_floor_list,
    "Facing": facing_list,
    "Balcony": balcony_list,
    "Page Number": page_numbers
})

# Convert numeric columns
df["Price"] = pd.to_numeric(df["Price"], errors="coerce")
df["Bathrooms"] = pd.to_numeric(df["Bathrooms"], errors="coerce")
df["Area (sqft)"] = pd.to_numeric(df["Area (sqft)"], errors="coerce")
df["BHK"] = pd.to_numeric(df["BHK"], errors="coerce")

print(df.head())
```
