```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import time
import re

# ---------------- LISTS ----------------

bhk_list = []
location_list = []
price_list = []
furnishing_list = []
bathroom_list = []
tenant_list = []
availability_list = []
area_list = []
area_type_list = []
floor_no_list = []
total_floor_list = []
facing_list = []
page_numbers = []
extra_charges_list = []

headers = {
    "content": "image/jpg",
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/145.0.0.0 Safari/537.36",
    "accept-encoding":"gzip, deflate, br, zstd"
    }
total_time = time.time()

for i in range(1, 51):   

    start_time = time.time()

    if i == 1:
        URL = "https://www.magicbricks.com/flats-for-rent-in-hyderabad-pppfr"
    else:
        URL = f"https://www.magicbricks.com/flats-for-rent-in-hyderabad-pppfr/page-{i}"

    page = requests.get(URL, headers=headers)
    soup = BeautifulSoup(page.text, "html.parser")

    cards = soup.find_all("div", class_="mb-srp__card")

    for card in cards:

        # ---------------- PRICE ----------------
        price_tag = card.find("div", class_="mb-srp__card__price--amount")
        if price_tag:
            price_text = price_tag.text.strip().replace("₹", "").replace(",", "").lower()
            if "call" in price_text:
                price = np.nan
            elif "lac" in price_text:
                value = re.search(r"([\d\.]+)", price_text)
                if value:
                    price = float(value.group(1)) * 100000
                else:
                    price = np.nan
            else:
                # Handle plain numbers (direct price in rupees)
                value = re.search(r"([\d\.]+)", price_text)
                if value:
                    price = float(value.group(1))
                else:
                    price = np.nan
        else:
            price = np.nan
        price_list.append(price)

        other_charges_tag = card.find("div", class_="mb-srp__card__price--other-charges")

        if other_charges_tag:
            other_text = other_charges_tag.text.strip().lower()

            if "security deposit" in other_text:
                extra_charge = "yes"
            elif "see other charges" in other_text:
                extra_charge = "no"
            else:
                extra_charge = "no"
        else:
            extra_charge = "no"

        extra_charges_list.append(extra_charge)
    
        # ---------------- TITLE (FOR BHK + LOCATION) ----------------
        title_tag = card.find("h2", class_="mb-srp__card--title")
        title = title_tag.text.strip() if title_tag else ""

        # Extract BHK
        bhk_match = re.search(r'(\d+)\s*BHK', title)
        bhk = bhk_match.group(1) if bhk_match else np.nan
        bhk_list.append(bhk)

        # Extract Location
        location = np.nan
        if " in " in title:
            location = title.split(" in ")[1].replace(", Hyderabad", "").strip()
        location_list.append(location)

        # ---------------- FEATURES ----------------
        furnishing = bathroom = tenant = availability = area = area_type = floor_no = total_floor = facing = np.nan

        summary_items = card.find_all("div", class_="mb-srp__card__summary__list--item")

        for item in summary_items:

            key = item.get("data-summary")
            value_tag = item.find("div", class_="mb-srp__card__summary--value")

            if not value_tag:
                continue

            value = value_tag.text.strip()

            if key == "furnishing":
                furnishing = value

            elif key == "bathroom":
                bathroom = value

            elif key == "tenent-preffered":
                tenant = value

            elif key == "status":
                availability = value

            elif key in ["super-area", "carpet-area"]:
                area = value.replace(" sqft", "")
                area_type = key

            elif key == "floor" and "out of" in value:
                parts = value.split(" out of ") 
                if parts[0] == "Ground":  #changed to handle Ground floor case
                    floor_no = 0
                else:
                    floor_no = parts[0]
                total_floor = parts[1]

            elif key == "facing":
                facing = value


        furnishing_list.append(furnishing)
        bathroom_list.append(bathroom)
        tenant_list.append(tenant)
        availability_list.append(availability)
        area_list.append(area)
        area_type_list.append(area_type)
        floor_no_list.append(floor_no)
        total_floor_list.append(total_floor)
        facing_list.append(facing)
        page_numbers.append(i)

    print(f"Page {i} completed in {time.time() - start_time:.2f} seconds")
    time.sleep(5)  # polite delay

print("Total Time:", time.time() - total_time)

# ---------------- CREATE DATAFRAME ----------------

df = pd.DataFrame({
    "Bedrooms": bhk_list,
    "Location": location_list,
    "Price": price_list,
    "Security Deposit": extra_charges_list,
    "Furnishing": furnishing_list,
    "Bathrooms": bathroom_list,
    "Tenant Preferred": tenant_list,
    "Availability": availability_list,
    "Area (sqft)": area_list,
    "Area Type": area_type_list,
    "Floor Number": floor_no_list,
    "Total Floors": total_floor_list,
    "Facing": facing_list,
    "Page Number": page_numbers
    
})

# ---------------- CLEAN NUMERIC COLUMNS ----------------

df["Price"] = pd.to_numeric(df["Price"], errors="coerce")
df["Bathrooms"] = pd.to_numeric(df["Bathrooms"], errors="coerce")
df["Area (sqft)"] = pd.to_numeric(df["Area (sqft)"], errors="coerce")
df["Bedrooms"] = pd.to_numeric(df["Bedrooms"], errors="coerce")
df["Floor Number"] = pd.to_numeric(df["Floor Number"], errors="coerce")
df["Total Floors"] = pd.to_numeric(df["Total Floors"], errors="coerce")
df = df.dropna(subset=["Price"])


print(df.head())
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import re
import time

titles = []
prices = []
bedrooms = []
locations = []
bathrooms = []
floors = []
areas = []
tenant_pref = []
furnishing = []
pagenum = []

total_time = time.time()

for i in range(1, 51):

    start_time = time.time()

    if i == 1:
        URL = "https://www.magicbricks.com/flats-for-rent-in-hyderabad-pppfr"
    else:
        URL = f"https://www.magicbricks.com/flats-for-rent-in-hyderabad-pppfr/page-{i}"

    headers = {
        "User-Agent": "Mozilla/5.0"
    }

    page = requests.get(URL, headers=headers)
    soup = BeautifulSoup(page.text, "html.parser")

    cards = soup.find_all("div", class_="mb-srp__card")

    for card in cards:

        # ---------------- PRICE ----------------
        price_tag = card.find("div", class_="mb-srp__card__price--amount")

        if price_tag:
            price_text = price_tag.text.strip()

            if "Call for Price" in price_text:
                price = np.nan
            else:
                price = re.sub(r"[^\d]", "", price_text)
        else:
            price = np.nan

        # ---------------- TITLE ----------------
        title_tag = card.find("h2", class_="mb-srp__card--title")

        if title_tag:
            title_text = title_tag.text.strip()

            # Bedrooms
            bed_match = re.search(r"(\d+)\s*BHK", title_text)
            bed = bed_match.group(1) if bed_match else np.nan

            # Location
            loc_match = re.search(r"in (.+)", title_text)
            loc = loc_match.group(1) if loc_match else np.nan
        else:
            bed = np.nan
            loc = np.nan

        # ---------------- SUMMARY LIST ----------------
        summary = card.find_all("div", class_="mb-srp__card__summary__list")

        bath = np.nan
        floor = np.nan
        area = np.nan
        tenant = np.nan
        furnish = np.nan

        for item in summary:
            text = item.text.strip()

            # Bathrooms
            if "Bathroom" in text:
                bath_match = re.search(r"\d+", text)
                if bath_match:
                    bath = bath_match.group()

            # Floor
            if "Floor" in text:
                if "Ground" in text:
                    floor = 0
                else:
                    floor_match = re.search(r"\d+", text)
                    if floor_match:
                        floor = floor_match.group()

            # Area
            if "sqft" in text:
                area_match = re.search(r"\d+", text)
                if area_match:
                    area = area_match.group()

            # Tenant Preferred
            if "Bachelors" in text or "Family" in text:
                tenant = text

            # Furnishing
            if "Furnished" in text or "Semi-Furnished" in text or "Unfurnished" in text:
                furnish = text

        prices.append(price)
        bedrooms.append(bed)
        locations.append(loc)
        bathrooms.append(bath)
        floors.append(floor)
        areas.append(area)
        tenant_pref.append(tenant)
        furnishing.append(furnish)
        pagenum.append(i)

    print(f"Page {i} completed in {time.time()-start_time:.2f} seconds")
    time.sleep(2)  # polite delay

print("Total Time Completed in seconds", str(time.time()-total_time))


# ---------------- CREATE DATAFRAME ----------------

df = pd.DataFrame({
    "Price": prices,
    "Bedrooms": bedrooms,
    "Location": locations,
    "Bathrooms": bathrooms,
    "Floor": floors,
    "Area_sqft": areas,
    "Tenant Preferred": tenant_pref,
    "Furnishing": furnishing,
    "Page Number": pagenum
})


# Convert numeric columns
df["Price"] = pd.to_numeric(df["Price"], errors="coerce")
df["Bedrooms"] = pd.to_numeric(df["Bedrooms"], errors="coerce")
df["Bathrooms"] = pd.to_numeric(df["Bathrooms"], errors="coerce")
df["Floor"] = pd.to_numeric(df["Floor"], errors="coerce")
df["Area_sqft"] = pd.to_numeric(df["Area_sqft"], errors="coerce")


# Remove Call for Price rows
df = df.dropna(subset=["Price"])
df
```
